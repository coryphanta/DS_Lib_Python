```python
import numpy as np

a = np.array([[1,6], [2,8], [3,11], [3,10], [1,7]])

tmp_x = []
tmp_y = []

for x,y in a:
    tmp_x.append(x)
    tmp_y.append(y)
    
mean_a = [[np.mean(tmp_x)], [np.mean(tmp_y)]]

print(mean_a)
```

```python
a_centered = np.subtract(a, np.resize(mean_a, (5, 2)))

print(a_centered)
```

```python
a_centered_sp = a_centered.T[0] @ a_centered.T[1]
print(a_centered_sp / (a_centered.shape[0]-1))
```
